```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL="https://www.flipkart.com/search?q=tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
#headerrs for request
HEADERS=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
          'Accept-Language':'en-US,en;q=0.5'})
```

```python
#http request
webpage=requests.get(URL,headers= HEADERS)
```

```python
webpage
```

```python
type(webpage.content)
```

```python
#soup object containing all data
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
#feching links as list og tag object
links=soup.find_all("a", attrs={'_1fQZEK'})
```

```python
link=links[0].get('href')
```

```python
product_list="https://www.flipkart.com" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
new_soup=BeautifulSoup(new_webpage.content,"html.parser")
```

```python
new_soup
```

```python
_title=new_soup.find_all("span", attrs={'B_NuCI'})
```

```python
for title in _title:
    print(title.text.strip())
```

```python
_price=new_soup.find_all("div", attrs={'_30jeq3 _16Jk6d'})
```

```python
for price in _price:
    print(price.text.strip())
```

```python
_rating=new_soup.find_all("div", attrs={'_3LWZlK'})
```

```python
for rating in _rating:
    i=rating.text.strip()
    print(i)
    break
```
